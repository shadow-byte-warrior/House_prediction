```python
columns_to_remove =['date','yr_renovated','street','city','statezip','country']
df =df.drop(columns=columns_to_remove)
```

```python
df.head(4)
```

```python

import scipy.stats as stats
```

```python
z_scores = stats.zscore(df)
threshold =3
print("size before removing outlier",df.shape)
outliers_df =df[(z_scores>threshold).any(axis=1)]
df =df[(z_scores<=threshold).all(axis=1)]
print("size after removing outlier",df.shape)
```

```python
outliers_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

scaler.fit(df)

df_scaler =pd.DataFrame(scaler.transform(df),columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x =df.drop('price',axis=1)
y=df['price']

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(f"X_train{X_train.shape}")
print(f"X_test{X_test.shape}")
```

```python
models =[
    ('Random Forest', RandomForestRegressor()),
    ('LinearRegression',LinearRegression()),
    ('DecisionTreeRegressor',DecisionTreeRegressor()),
    ('KNeighborsRegressor', KNeighborsRegressor())
]
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
for name,model in models:
    print(name)
    print()
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    print("mean_squared_error", mean_absolute_error(y_test,y_pred))
    print('\n')
   
```
